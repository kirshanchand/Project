### __Goal__<br>
How do economic, health, and digital factors influence the well being and development of countries worldwide?

- Do countries with higher education and health expenditure have better life expectancy?
- Are countries with better digital infrastructure also investing more in education or health?
- Did unemployment rise in countries with high inflation or economic instability?
- Which top 10 countries have shown the fastest GDP growth over the past decade?
- Which top 10 countries perform well across health, education, economy, and digitalization?


```python
# Imports libraries
import pandas as pd 
import numpy as np
import plotly.express as px 
import plotly.graph_objects as go
from great_tables import (
    GT, md, style, google_font, loc
)
```

```python
# Dataset link: https://www.kaggle.com/datasets/yogafsyahputra/worldbank-development
# Load the dataset into a pandas dataframe  
df = pd.read_csv('worldbank_development_indicators.csv')
df
```

```python
# Show the info about dataset 
df.info()
```

```python
# Show the descriptive statistics of the data
df.describe()
```

```python
# Get all the columns name to understand dataset structure 
df.columns
```

```python
#  Rename long and unclear column names to more readable and analysis friendly versions
df = df.rename(columns={
    'GDP_per_capita_USD': 'GDP Per Capita (USD)',
    'Inflation_annual_pct': 'Inflation',
    'Life_expectancy_years': 'Life Expectancy',
    'Education_expenditure_pct_GDP': 'Education Expenditure',
    'Health_expenditure_pct_GDP': 'Health Expenditure',
    'Unemployment_pct': 'Unemployment',
    'Internet_users_pct': 'Internet Users (%)'
})
df.head()
```

##### 1. Do countries with higher education & health expenditure have better life expectancy?

```python
# Filter years 2000–2022
df = df[(df['Year'] >= 2000) & (df['Year'] <= 2022)]

# most recent data for each country
latest = df.sort_values("Year").groupby("Country").tail(1).reset_index(drop=True)
df1 = latest

# Required columns to use
cols = ['Country', 'Education Expenditure', 'Health Expenditure', 'Life Expectancy']

# Combined expenditure
df1['Total Expenditure'] = df1['Education Expenditure'] + df1['Health Expenditure']
```

```python
# Scatter plot
fig = px.scatter(
    df1,
    x='Total Expenditure',
    y='Life Expectancy',
    hover_name='Country',
    trendline='ols',
    title='<b>Investing in Education and Health Correlates with Higher Life Expectancy',
    subtitle = 'Countries allocating more than 15% of GDP to social spending consistently achieve life expectancies<br>above 80 years, while low spending nations show significantly lower longevity outcomes.',
    labels={
        'Total Expenditure':'Education + Health Expenditure',
        'Life Expectancy':'Life Expectancy (years)'
    },
    height=500,
    width=700
)
# Trendline 
fig.data[-1].line.color = 'red'
# Title styling
fig.update_layout(
    title_font_family="Helvetica",
    title_font_size=18,
    font=dict(family='Helvetica', size=14, color="#1A1A1A"),
    margin = {'t':150}
)
fig.show()
```

##### 2. Are countries with better digital infrastructure also investing more in education or health?


```python
# Select required columns 
cols = [
    'Country',
    'Year',
    'Education Expenditure',
    'Health Expenditure',
    'Internet Users (%)'
]

df_clean = df[cols].copy()

# cleaning dataset
df_clean = df_clean.dropna() 
df_clean = df_clean.drop_duplicates()

# Keep latest record for each country 
df_latest = (
    df_clean.sort_values('Year')        
           .groupby('Country')          
           .tail(1)                     
           .reset_index(drop=True)
)

# Select top 10 countries with highest Internet Users 
top10 = (
    df_latest.sort_values('Internet Users (%)', ascending=False)
             .head(10)
             .loc[:, ['Country', 'Education Expenditure', 'Health Expenditure', 'Internet Users (%)']]
)

# Rename specific country name (UAE)
top10['Country'] = top10['Country'].replace({'United Arab Emirates': 'UAE'})
# Indexing for Table 
top10.index = range(1, len(top10)+1)
top10_reset = top10.reset_index(drop=True)
top10
```

```python
# Table and heatmap
# Get min and max values for heatmap
minv = top10_reset['Internet Users (%)'].min()
maxv = top10_reset['Internet Users (%)'].max()

# Create GT table with Internet Users highlighted in purple
(
    GT(top10_reset)
    .tab_header(
        title=md("<b>Digital Infrastructure Correlates with Higher Social Spending"),
        subtitle=md("Countries with higher internet penetration consistently allocate more funding to both education and health,<br>confirming a strong link between digital connectivity and social investment.")
    )
    # Dataset source name
    .tab_source_note('Source: World Bank Development Indicators')
    .opt_align_table_header('left')
    # Title styling
    .tab_options(
        heading_title_font_size='22px',
        heading_subtitle_font_size='16px',
        column_labels_font_size='16px',
        table_font_size='16px'
    )
    .tab_style(
        style = style.text(font = google_font(name = "Helvetica")), 
        locations = loc.title()
    )
    # Heatmap
    .data_color(
        columns='Internet Users (%)',   
        domain=[minv, maxv],         
        palette='Purples'              
    )
)
```

##### 3. Did unemployment rise in countries with high inflation or economic instability?

```python
# Select relevant columns and remove duplicates
df_inf = df[['Country', 'Year', 'Inflation', 'Unemployment']].drop_duplicates()

# Identify the most affected country based on average Inflation + Unemployment
country_impact = df_inf.groupby('Country')[['Inflation', 'Unemployment']].mean().sum(axis=1)
most_affected = country_impact.idxmax()

# Filter data for the most affected country and sort by year
df_top = df_inf[df_inf['Country'] == most_affected].sort_values('Year')

# Prepare full timeline values (2000–2022) for Inflation and Unemployment
years_full = list(range(2000, 2023))
inflation_vals = df_top.set_index('Year').reindex(years_full)['Inflation'].tolist()
unemp_vals = df_top.set_index('Year').reindex(years_full)['Unemployment'].tolist()
```

```python
# line chart
fig = go.Figure()

# Inflation trace
fig.add_trace(go.Scatter(
    x=years_full,
    y=inflation_vals,
    mode='lines',
    name=f'{most_affected} – Inflation',
    line=dict(width=4, color='#d62728')
))

# Unemployment trace
fig.add_trace(go.Scatter(
    x=years_full,
    y=unemp_vals,
    mode='lines',
    name=f'{most_affected} – Unemployment',
    line=dict(width=4, color='#1191ed', dash='dot'),
    yaxis='y2'
))

# Title and subtitle
fig.update_layout(
    title=(
        f"<b>Rising Inflation and Unemployment in Zimbabwe: Trends Over 2000–2022</b><br>"
        "<span style='font-size: 14px'>In this chart, Zimbabwe’s inflation (Red) surged sharply after 2008, while unemployment (Blue) also rose, highlighting<br>the country’s economic instability."
    ),
    title_font_size=18,
    font=dict(family='Helvetica', size=16),
    width=800,
    height=500,
    xaxis=dict(title='Year'),
    yaxis=dict(title='Inflation (%)', showgrid=False),
    yaxis2=dict(title='Unemployment (%)', overlaying='y', side='right', showgrid=False),
    legend_title="Indicators",
    margin={'t':140},
    showlegend=False
)
fig.show()
```

##### 4. Which countries have shown the fastest GDP growth over the past decade?

```python
# Select required columns 
df1 = df[['Country', 'Year', 'GDP Per Capita (USD)']].dropna()

# Keep only last 10 years 
latest = df1['Year'].max()
df1 = df1[df1['Year'] >= latest - 10]

# Calculate GDP growth 
growth = (
    df1.sort_values(['Country', 'Year'])
       .groupby('Country')
       .agg(start_gdp=('GDP Per Capita (USD)', 'first'),
            end_gdp=('GDP Per Capita (USD)', 'last'))
)

# Percentage growth
growth['GDP Growth (%)'] = (growth['end_gdp'] - growth['start_gdp']) / growth['start_gdp'] * 100

# Top 10 fastest growing
top10_growth = growth['GDP Growth (%)'].nlargest(10).reset_index()
top10_growth
```

```python
# Bar chart
# Create 'flag' column to mark the country with highest GDP growth
top10_growth['flag'] = (top10_growth['GDP Growth (%)'] == top10_growth['GDP Growth (%)'].max()).astype(str)
# format GDP growth values to 2 decimal places
top10_growth['GDP Growth Label'] = top10_growth['GDP Growth (%)'].apply(lambda x: f"{x:.2f}")

# Horizontal bar chart
fig = px.bar(
    top10_growth.sort_values(by='GDP Growth (%)', ascending=True),
    x = 'GDP Growth (%)', 
    y = 'Country', 
    template='simple_white', 
    text='GDP Growth Label',
    title = '<b>Top 10 Countries with the Fastest GDP Growth (Last 10 Years)', 
    subtitle='This chart shows GDP per capita growth over the last decade, with Guyana at 229%, Bangladesh at 215%,<br>and Timor-Leste at 129%, highlighting the fastest growing economies.',
    height = 500, 
    width = 750,
    color='flag',  
    color_discrete_sequence=['lightgray', '#1191ed']    
)
# Remove the x and y ticks labels
fig.update_layout(yaxis_ticks="")
fig.update_layout(xaxis_ticks="")

# Title and subtitle styling 
fig.update_layout(
    font=dict(size=16, color="#1A1A1A", family="Helvetica"),
    title_font_family = 'Helvetica', 
    title_font_size = 20, 
    title_subtitle_font_size=15,
    margin = {'t':140},
    showlegend=False
)
# Add thin black border to bars
fig.update_traces(marker_line_width=0.5, marker_line_color="black")
# Extend x-axis a bit beyond max value
fig.update_xaxes(range=[0, top10_growth['GDP Growth (%)'].max() * 1.1])
fig.show()
```

##### 5. Which top 10 countries perform well across health, education, economy, and digitalization?

```python
# Columns for scoring
score_cols = [
    'GDP Per Capita (USD)',
    'Education Expenditure',
    'Health Expenditure',
    'Internet Users (%)'
]

# Preparing data
latest = df.sort_values("Year").groupby("Country").tail(1).reset_index(drop=True)
score = latest.dropna(subset=score_cols).set_index('Country')[score_cols]
score_norm = (score - score.min()) / (score.max() - score.min())
score_norm['Composite'] = score_norm.mean(axis=1)

# Top 10 by composite score
top10_comp = score_norm.nlargest(10, 'Composite').reset_index()

# Melt for facet bar chart
top10 = top10_comp.melt(
    id_vars='Country',
    value_vars=score_cols,
    var_name='Indicator',
    value_name='Normalized Score'
)
```

```python
# Facet bar chart
fig = px.bar(
    top10,
    x='Normalized Score',
    y='Country',
    facet_col='Indicator',
    orientation='h',
    color='Indicator',
    color_discrete_sequence=["#1191ed", "#2ca02c", "#d62728", "#b274ec"],
    title='<b>Top 10 Balanced Countries Across Economy, Health, Education & Digitalization',
    subtitle = 'This chart compares the top 10 countries based on normalized scores across GDP per capita, Education Exp, Health Exp, and internet usage.<br>It highlights which countries are most balanced and advanced across these key development indicators.',
    width=950,
    height=500
)

# Layout adjustments
fig.update_layout(
    title_font_family="Helvetica",
    title_font_size=20,
    title_subtitle_font_size = 14,
    font=dict(family='Helvetica', size=14, color="#1A1A1A"),
    showlegend=False,
    margin={'t':170}
)

# Update all annotation labels (font size and color)
fig.for_each_annotation(lambda a: a.update(font=dict(size=12, color="#1A1A1A")))
# Remove all individual x-axis label and use a single clean label for the entire figure
fig.for_each_xaxis(lambda x: x.update(title_text=None))

# y-axis label
fig.update_layout(yaxis_title="Country")

# X-axis label styling
fig.add_annotation(
    text="Normalized Score (0–1)",
    xref="paper",
    yref="paper",
    x=0.5,
    y=-0.20,
    showarrow=False,
    font=dict(size=16, color="#1A1A1A", family="Helvetica")
)
fig.show()
```
